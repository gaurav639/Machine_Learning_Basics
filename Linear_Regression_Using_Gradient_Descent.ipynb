```python
import numpy as np
import pandas as pd
import requests
import io
import matplotlib.pyplot as plt
```

```python
# Here we are just taking the csv file directly from github
url = "https://raw.githubusercontent.com/chasinginfinity/ml-from-scratch/master/01%20Linear%20Regression%20using%20Least%20Squares/data.csv" 
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')))
```

```python
print(df.head())
```

```python
x = df.to_numpy()[:,0]
actual_y = df.to_numpy()[:,1]
plt.rcParams['figure.figsize'] = (12.0, 9.0)
plt.scatter(x, actual_y)
plt.show()
```

```python
m = float(len(x))
alpha = 0.0001
w = 0
b = 0

for i in range(1000):
  pred_y = w*x + b
  d_w = (1/m) * sum(x * (pred_y - actual_y))
  d_b = (1/m) * sum(pred_y - actual_y)
  w = w - alpha * (d_w)
  b = b - alpha * (d_b)

print(w,b)
```

```python
pred_y = w*x + b
plt.scatter(x,actual_y)
plt.plot([min(x), max(x)], [min(pred_y), max(pred_y)], color='red')
plt.show()
```
