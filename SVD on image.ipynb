```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
```

```python
path = 'https://upload.wikimedia.org/wikipedia/commons/d/d3/Albert_Einstein_Head.jpg'
A = io.imread(path)
```

```python
img = plt.imshow(A)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U,S,VT = np.linalg.svd(A, full_matrices = False)
S = np.diag(S)
```

```python
S.shape
```

Now we have to compress it to 30% of its singular values. i.e., 0.3 * 3250 = 975 singular values only we have to take.

```python
r = 975
A_compressed = U[:, :r] @ S[:r,:r] @ VT[:r,:]
img_compressed = plt.imshow(A_compressed)
img_compressed.set_cmap('gray')
plt.axis('off')
plt.show()
```

**So, even after compressing the image to 30% singular values, we are getting a similar image without having any major loss.**
